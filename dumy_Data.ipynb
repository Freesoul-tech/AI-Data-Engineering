```python
# Creating dumy data with Louis Jeremiah

import numpy as np

#  seed value
np.random.seed(0)

# Mean and cov
mean = [-3, 0]
cov = [[1.0, 0.8], [0.8, 1.0]]
data1 = np.random.multivariate_normal(mean, cov, 500)

import matplotlib.pyplot as plt

plt.scatter(data1[:, 0], data1[:, 1], alpha=0.5)
plt.title('Scatter Plot of Problem 1 Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 5))

# Histogram
plt.subplot(1, 2, 1)
plt.hist(data1[:, 0], bins=30, alpha=0.7)
plt.title('Histogram for X Dimension')
plt.xlabel('X')
plt.ylabel('Frequency')
plt.xlim(-6, 6)

# Histogram for Y
plt.subplot(1, 2, 2)
plt.hist(data1[:, 1], bins=30, alpha=0.7)
plt.title('Histogram for Y Dimension')
plt.xlabel('Y')
plt.ylabel('Frequency')
plt.xlim(-6, 6)

plt.tight_layout()
plt.show()

# New mean
mean2 = [0, -3]

# Random numbers
data2 = np.random.multivariate_normal(mean2, cov, 500)

# Scatter plot for both data sets
plt.scatter(data1[:, 0], data1[:, 1], alpha=0.5, label='Problem 1 Data')
plt.scatter(data2[:, 0], data2[:, 1], alpha=0.5, label='Problem 4 Data', color='r')
plt.title('Scatter Plot of Combined Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
combined_data = np.vstack((data1, data2))
print(combined_data.shape)
labels = np.hstack((np.zeros(500), np.ones(500)))
final_data = np.column_stack((combined_data, labels))
print(final_data.shape)  # Should print (1000, 3)




```
